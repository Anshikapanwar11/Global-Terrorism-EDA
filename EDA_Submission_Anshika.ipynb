Ultimately, this project translates historical data into a strategic roadmap. By identifying concentrated geographic hotspots and the most lethal, highly-successful attack methodologies, the insights generated here provide stakeholders with a clear mandate for resource allocation. ## ***1. Know Your Data***

### Import Libraries

````python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')
````

### Dataset Loading

````python
# Load the dataset
print("Loading UNGTA dataset...")
df = pd.read_csv('Global Terrorism Data.csv', encoding='ISO-8859-1', low_memory=False)
````

### Dataset First View

````python
# Dataset First Look
# 1. Check the shape of the dataset (Rows and Columns)
print(f"Dataset Shape: {df.shape[0]} rows and {df.shape[1]} columns.\n")

# 2. View the first 5 rows to understand the structure
display(df.head())

# 3. Get basic information about data types and non-null counts
display(df.info())

# 4. Get statistical summary of numerical columns
display(df.describe())
````

## ***2. Understanding Your Variables***

````python
# Print all column names to understand available variables
print("All available columns in the raw dataset:")
print(list(df.columns))

# Check for the percentage of missing values in each column
print("\nPercentage of missing values per column:")
missing_percentages = (df.isnull().sum() / len(df)) * 100
display(missing_percentages.sort_values(ascending=False).head(20)) # Show top 20 worst offenders

# Focus variables for our analysis:
# iyear (Year), region_txt (Region), country_txt (Country)
# attacktype1_txt (Attack Type), weaptype1_txt (Weapon), targtype1_txt (Target)
# nkill (Kills), nwound (Wounds), success (Attack Success Rate)
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# 1. Create a copy to preserve the original raw dataframe
df_clean = df.copy()

# 2. Keep only the columns relevant for our Business Objective
# (This drops over 100 useless columns and saves massive amounts of memory)
cols_to_keep = [
    'iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'provstate', 'city',
    'latitude', 'longitude', 'attacktype1_txt', 'targtype1_txt',
    'gname', 'weaptype1_txt', 'nkill', 'nwound', 'success'
]
df_clean = df_clean[cols_to_keep]

# 3. Rename columns so they are easier to read in our visualizations
df_clean.rename(columns={
    'iyear':'Year', 'imonth':'Month', 'iday':'Day',
    'country_txt':'Country', 'region_txt':'Region', 'attacktype1_txt':'Attack_Type',
    'targtype1_txt':'Target_Type', 'gname':'Group', 'weaptype1_txt':'Weapon_Type',
    'nkill':'Kills', 'nwound':'Wounds', 'success':'Success'
}, inplace=True)

# 4. Handle Missing Values (Crucial Step)
# Assuming that if kills/wounds are not reported, they are 0
df_clean['Kills'] = df_clean['Kills'].fillna(0)
df_clean['Wounds'] = df_clean['Wounds'].fillna(0)

# Fill missing location data with 'Unknown'
df_clean['provstate'] = df_clean['provstate'].fillna('Unknown')
df_clean['city'] = df_clean['city'].fillna('Unknown')

# 5. Feature Engineering
# Create a new column 'Total_Casualties' to measure overall human impact
df_clean['Total_Casualties'] = df_clean['Kills'] + df_clean['Wounds']

print(f"Data Wrangling Complete!")
print(f"Cleaned Dataset Shape: {df_clean.shape[0]} rows and {df_clean.shape[1]} columns.")
display(df_clean.head())
````

#### Chart - 1

````python
# Chart - 1 visualization code
try:
    plt.figure(figsize=(15, 6))
    sns.histplot(data=df_clean, x='Year', bins=len(df_clean['Year'].unique()), kde=True, color='darkred')
    plt.title('Global Terrorist Attacks per Year (1970 - 2017)', fontsize=16)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Number of Incidents', fontsize=12)
    plt.show()
except Exception as e:
    print(f"Error generating chart: {e}")
````

Hostage Taking, Hijacking, and Assassinations have incredibly high success rates (often above 80%), whereas unarmed assaults have a notably lower probability of success.
